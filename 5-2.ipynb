## 검증 세트 
- 

```python
import pandas as pd 

wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.head()
```

```python
# 1. 훈련용, 채점용 만들기

data = wine[['alcohol', 'sugar', 'pH']]
target = wine[['class']]
```

```python
# 2. 훈련용 , 채점용 넣고 섞어주기

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(data, target)
```

```python
##  3. 검증 세트 val 만들기

sub_input, val_input, sub_target, val_target = \
train_test_split(train_input, train_target)
```

```python
print(sub_input.shape)
print(val_input.shape)
```

```python
# 4. 디시젼 트리 (결정 트리)       ----> 검증세트 채점

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(sub_input, sub_target)

print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
```

```python
# 5. 테스트 세트 채점

print(dt.score(test_input, test_target))
```

## 교차 검증 
- from sklearn.model_selection import cross_validate

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt,train_input,train_target)          # dt를 교차검증 --> test_score 확인 : 80점대 중반으로 균일하게 나옴.
print(scores)
```

## 그리드 서치 
- from sklearn.model_selection import GridSearchCV
- 하이퍼파라미터 탐색, 교차 검증을 한번에 수행 = 별도로 cross_validate 함수 불러올 필요 없음.

```python
from sklearn.model_selection import GridSearchCV

params = {
    'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004]           #하이퍼 파라미터 : 파라미터값 직접 지정
}
```

```python
gs = GridSearchCV(dt, params, n_jobs = -1)            #n_jobs = -1 : 시스템에 있는 모든 코어값 활용 
```

```python
gs.fit(train_input,train_target)
```

```python
dt = gs.best_estimator_
print(gs.score(train_input, train_target))
```
