## 03. 차원축소 알고리즘
- 머신러닝에서는 특성을 차원이라고 부름 --> 저장용량을 줄이기 위해서 차원축소

### 1. 주성분 분석
`from sklearn.decomposition import PCA`
- (2차원을 1차원으로 만드는) 두개의 특성을 하나의 특성으로 만드는 과정  = 여러 개의 독립변수들을 잘 설명해줄 수 있는 주된 성분을 추출
- 분산 (데이터가 널리 퍼져있는 정도)의 큰 방향 = 원본의 특성을 가장 잘 나타냄
    - 주성분 = 원본 특성의 개수만큼 찾음.

```python
import numpy as np
fruits = np.load('data/fruits_300.npy')
fruits_2d = fruits.reshape(-1, 100*100)  # 2차원으로 reshape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=50)                # 300개에서  -> 주성분의 특징을 가진 50개로 압축
pca.fit(fruits_2d)                        # 비지도 학습이므로 타깃값 지정 x
```

```python
# components_ : 주성분 저장 

pca.components_.shape
```

```python
# (50,10000) 그림 그려주기

import matplotlib.pyplot as plt

def draw_fruits(arr, ratio=1):
    n = len(arr)

    rows = int(np.ceil(n/10))
    cols = n if rows < 2 else 10

    fig, axs = plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)

    for i in range(rows):
        for j in range(cols):
            if i*10 + j < n:
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()
```

```python
draw_fruits(pca.components_.reshape(-1, 100,100))
```

```python
# transform() : 원본 데이터 차원을 50으로 줄이기 

fruits_pca = pca.transform(fruits_2d)
fruits_pca.shape
```

```python
# inverse_transform : 원본 데이터(10000)를 복원해서 그리기

fruits_inverse = pca.inverse_transform(fruits_pca)
fruits_inverse.reshape(-1, 100, 100)                    # 3d 원본으로 복원
draw_fruits(fruits[200:300])
```

```python
# explained_variance_ratio_ : 설명된 분산 (주성분이 원본데이터의 분산을 얼마나 잘 나타내는지 확인)
np.sum(pca.explained_variance_ratio_)              # 92%가 넘는 분산을 유지
# pca.explained_variance_ratio_                       # 초반 10개가 대부분의 분산을 띠고 있음.
```

#### # 로지스틱 회귀 : 지도학습 (타겟 필요)

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

target = np.array([0]*100 + [1]*100 + [2]*100)
```

#### # 교차검증 (원본 데이터와 주성분으로 축소한 경우 걸리는 시간 비교) == 주성분 축소시 훈련속도 향상

```python
# 원본 데이터 (fruits_2d)
from sklearn.model_selection import cross_validate
scores = cross_validate(lr, fruits_2d, target)
print(scores)
```

```python
# 주성분 축소 
scores = cross_validate(lr, fruits_pca, target)
print(scores)
```

#### # 설명된 분산의 50%에 달하는 주성분 개수 찾기

```python
pca = PCA(n_components = 0.5)       
pca.fit(fruits_2d)

pca.n_components_                  # 단 2개의 특성만으로 원본데이터의 있는 분산의 50% 표현 가능
```

```python
fruits_pca = pca.transform(fruits_2d)        # 주성분으로 원본데이터의 차원 변경 
fruits_pca.shape
```

```python
scores = cross_validate(lr, fruits_pca, target)
scores
```

#### # k-평균 군집 알고리즘

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)         #random_state=42 잘 분류되는 형태
km.fit(fruits_pca)
```

```python
km.labels_ # 배열의 값을 labels_에 [0,1,2]로 저장해둠.
```

```python
print(np.unique(km.labels_, return_counts = True))     # 몇개로 나누어들어가는지 알 수 있음. 
```

```python
# 그래프로 그리기 

for label in range(3):
    data = fruits_pca[km.labels_ == label]   
    plt.scatter(data[:,0], data[:,1])               # 0, 1인덱스는 각각 x축, y축을 의미

plt.legend(['apple', 'pineapple', 'banana'])
plt.show()
```
