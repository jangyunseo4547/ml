## 01. 군집 알고리즘

#### 연습 1) 과일 군집 나누기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits = np.load('data/fruits_300.npy')
```

```python
fruits.shape                           # 배열의 길이, 모양 (3차원) : 300장의 사진이 100 * 100 크기로 들어있음.
```

```python
fruits[0, 0, :] # 첫번째 이미지의 첫번째 행의 처음 ~ 끝까지 (100개)
```

```python
plt.imshow(fruits[0], cmap = 'gray')        # cmap : 0에 가까울수록 검게 / 높은 값은 밝게 나타남.
plt.show()
```

```python
plt.imshow(fruits[0], cmap = 'gray_r')        # 우리눈에 보기 좋도록 다시 출력 ==> 짙을수록 255에 가까움 / 밝은 부분은 0에 가까움.
plt.show()
```

```python
fig, axs = plt.subplots(1,2)       # subplots(,) : 그래프를 쌓을 1행과 2열 지정
axs[0].imshow(fruits[100])
axs[1].imshow(fruits[200])
plt.show()
```

```python
# 1. 픽셀값 분석하기 : 100 * 100 이미지를 펼쳐 길이가 10,000인 1차원 배열(배열 계산에 편리)로 만들기

apple = fruits[0:100].reshape(-1, 100*100)               # -1: 자동으로 남은 차원 할당 / 2,3차원을 10000으로 합침 
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
```

```python
apple.shape      
```

```python
# 사과 샘플 100개에 대한 픽셀 평균값 계산

apple.mean(axis = 1)   # axis : 1차원인경우 1(가로), 0(세로) / 2차원인 경우 행 (0), 열(1) 의미
```

```python
## 1-1 히스토그램

plt.hist(np.mean(apple, axis = 1))
plt.hist(np.mean(pineapple, axis = 1))
plt.hist(np.mean(banana, axis = 1))

plt.legend(['apple', 'pineapple', 'banana'])        # 항목 표시
plt.show()                                          # 바나나와 사과, 파인애플은 극명한 차이를 보임 (숫자가 채워져 있는지 없는지)
```

#### 해석 
- 바나나는 픽셀 평균이 40이하에 많이 분포되어 있어 사과, 파인애플의 픽셀 평균과 큰 차이를 보임 (분류 가능)
- 그러나 파인애플, 사과는 샘플의 평균만으로는 분류 어려움. -> 픽셀의 평균값 구분 필요 ! 

```python
## 1-2 막대 그래프 / axis = 0 (픽셀의 평균 계산)

fig, axs = plt.subplots(1,3, figsize = (20,5))
axs[0].bar(range(10000), np.mean(apple, axis = 0))
axs[1].bar(range(10000), np.mean(pineapple, axis = 0))
axs[2].bar(range(10000), np.mean(banana, axis = 0))

plt.show()
```

#### # 결과 해석 
- 과일마다 값이 높은 구간이 다름.
- 사과 : 사진 아래쪽으로 갈수록 값이 높아짐 / 파인애플 : 비교적 고르면서 전체 평균이 높음. / 바나나 : 중앙의 픽셀값이 높음.

```python
## 1-3 이미지로 출력 

apple_mean = np.mean(apple, axis=0).reshape(100,100)          # reshape : 다시 사진형태로 만들기 위해서 
pineapple_mean = np.mean(pineapple, axis=0).reshape(100,100)
banana_mean = np.mean(banana, axis=0).reshape(100,100)

fig, axs = plt.subplots(1,3, figsize = (20,5))            # 1행 3열로 그리기
axs[0].imshow(apple_mean, cmap = 'gray_r')
axs[1].imshow(pineapple_mean, cmap = 'gray_r')
axs[2].imshow(banana_mean, cmap = 'gray_r')
plt.show()
```

```python
# 2. 평균값과 가까운 사진 고르기

abs_diff = np.abs(fruits - apple_mean)       # abs() : 넘파이에서의 절댓값 계산 함수 
print(abs_diff)                                   # abs_diff : (300,100,100) 크기의 3차원 배열
abs_mean = np.mean(abs_diff, axis = (1,2))        # abs_mean : 각 샘플의 오차의 평균 (300, )인 1차원 배열
print(abs_mean.shape)
```

```python
apple_index = np.argsort(abs_mean)[0:100]         #argsort : 넘파이에 있는 sort / 오차의 평균과 가장 조금 차이나는 100개를 정렬함.

fig, axs = plt.subplots(10,10, figsize = (10,10))   # 전체 화이트보드의 사이즈를 10행 10열로

# 이중 for문을 활용한 100개의 이미지 출력
for i in range(10):                                 
    for j in range(10):
        axs[i,j].imshow(fruits[apple_index[i*10 + j]])       # i가 1행일때 j는 0~9열까지 반복이 돌아감  -> i가 9까지 반복

plt.show()                                          


# 예를 들어 i = 9, j =9일때, apple_index[99]번째값을 가져옴. -> i,j가 각각 0~9 돌때까지 반복 
```
