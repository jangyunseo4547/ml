### 2. 새로운 특성 만들기
- 가족으로 묶기 (부모자식, 형제자매)
- age 그룹
- fare 그룹

```python
# 1. Family_size

train_data['Family_Size'] = train_data['SibSp'] + train_data['Parch'] + 1   # 자기 자신 포함
train_data['Family_Size'].value_counts()
```

```python
# 2. Age를 나이대별로 그룹화하는 함수
def age_group(age):
    if age <= 12:
        return 'Child'
    elif age <= 18:
        return 'Teenager'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

train_data['AgeGroup'] = train_data['Age'].apply(age_group)

print(train_data['AgeGroup'].value_counts())
```

```python
# 3. Fare를 요금별로 그룹화하는 함수

def fare_group(fare):
    if fare <= 10:
        return 'Low'
    elif fare <= 30:
        return 'Medium'
    else:
        return 'High'

train_data['FareGroup'] = train_data['Fare'].apply(fare_group)

print(train_data['FareGroup'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
train_data['AgeGroup'] = le.fit_transform(train_data['AgeGroup'])
```

```python
train_data['FareGroup'] = le.fit_transform(train_data['FareGroup'])
```

```python
train_data.head() # high = 0, low =1 , medium =2
```

## 3. 랜덤 포레스트

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = \
train_test_split(train_data, target_data)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs = -1)   
```

```python
rf.fit(train_input, train_target)
```

```python
print(rf.score(train_input,train_target))
print(rf.score(test_input,test_target))
```

```python
rf = RandomForestClassifier(oob_score = True, n_jobs = -1, max_depth = 10)

rf.fit(train_data, target_data)    
```
